```python
from urllib.request import urlopen
import json
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
```

```python
fips_codes = {
    'Nassau': '12089',
    'St. Johns': '12109',
    'Baker': '12003',
    'Clay': '12019',
    'Duval': '12031'
    }

covid_values = {
    'Duval': [287, 18, 3, 1, 1, 1, 1, 1],
    'Clay': [32, 17, 13, 7],
    'Nassau': [11,6,1,1],
    'St. Johns': [36, 28, 22, 17, 8, 5],
    'Baker': [11]
}
```

```python
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)
```

```python
county_list = counties["features"]
# fips_list = list(fips_codes.values())
jacks_counties = [county for county in county_list if county['id'] in fips_codes.values()]
covid_count = []
fips = []
for each in fips_codes.keys():
    fips.append(fips_codes[each])
    covid_count.append(sum(covid_values[each]))
print(covid_count)
print(fips)

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/fips-unemp-16.csv",
                   dtype={"fips": str})

colorscale = [
    'rgb(193, 193, 193)',
    'rgb(239,239,239)',
    'rgb(195, 196, 222)',
    'rgb(144,148,194)',
    'rgb(101,104,168)',
    'rgb(65, 53, 132)'
]

fig = ff.create_choropleth(
    fips=fips, values=covid_count, scope=['FL'],
    binning_endpoints=[10, 50, 100, 200, 500], colorscale=colorscale,
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='Cases by County', title='Jacksonville COVID-19 Cases'
)
fig.layout.template = None
fig.show()
```
